Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
def people():

    number = str(input("Введите номер документа "))
    name = ""
    
    for numb in documents:
        if number in numb.values():
            name = numb['name']         
            break
        else:
            name = "Документ не найден"
            
    return name
    
def shelf():
    number = str(input("Введите номер документа "))
    shelf = ""
    for numb, doc in directories.items():
        if number in doc:
            shelf = numb
            break
        else:
            shelf = "Документ не найден"        
            
    return shelf 

def info():
    for info in documents:
        print('\033[1m' + f'{info["type"]} "{info["number"]}" {info["name"]}' + '\033[0m')
    
def add_shelf():
    number = str(input("Введите номер полки "))
    result = " "
    if number not in directories.keys():
        directories[number] = []
        result = f"Полка с номером {number} создана в директории"
        return result
    else:
        result = "Полка с таким номером уже существует"
        return result
  
                
def del_info():
    number = str(input("Введите номер документа "))
    result = ""
    element_index = 0
    
    for numb, doc in directories.items():
        if number in doc:
            doc.remove(number)
    
    for info in documents:
        if number in info['number']:
            element_index = documents.index(info)
            del(documents[element_index])
            result = "Документ удален"
            break
            
        else:
            result = "Документ с таким номером не найден"
                
    return result

def move_shelf():
    doc_number = str(input("Введите номер документа "))
    shelf_number = str(input("Введите номер полки "))
    
    result = " "
    
    for numb, doc in directories.items():
        if shelf_number in directories.keys() and doc_number in doc:
            doc.remove(doc_number)
            directories[shelf_number] += {doc_number}
            result = f"Документ {doc_number} пермещен в полку {shelf_number}"
            break
        elif shelf_number not in directories.keys():
            result = "Полка c таким номером не существует"
        else:
            result = "Документ не найден"
                   
    return result

def add_new():
    doc_type = str(input("Введите тип документа "))
    doc_number = str(input("Введите номер документа "))
    name = str(input("Введите ФИО владельца "))
    shelf_number = str(input("Введите номер полки "))
    result = " "
    
    new_dict = {'type': doc_type, 'number': doc_number, 'name': name}
    
    if shelf_number in directories.keys():
        documents.append(new_dict)
        directories[shelf_number] += {doc_number}
        result = "Документ создан и добавлен в директорию"
        
    else:
        result = "Полка не найдена"
    return result


def main():
    while True:
        command = str(input("Введите команду (для выхода введите 'exit') "))
        if command ==("p"):
            print(people())
        elif command ==("s"):
            print(shelf())
        elif command ==("l"):
            info()
        elif command ==("as"):
            print(add_shelf())
        elif command ==("d"):
            print(del_info())
        elif command ==("m"):
            print(move_shelf())
        elif command ==("a"):
            print(add_new())
        elif command ==("exit"):
            break
            
# print(directories)           
```

```python
main()
```
