```python
# a) f(x) = cos(x)^2
x1 = np.linspace(-5.0, 5.0, 100)
f1 = (np.cos(x1))**2
f1min = -f1
max1, _ = find_peaks(f1, prominence=0)
min1, _ = find_peaks(f1min, prominence=0)
plt.plot(f1)
plt.plot(min1, f1min[min1], "p")
plt.plot(max1, f1[max1], "x")
plt.show()
```

```python
#b) f(x) = x^3 - 9x
def f(x2):
    return x2**3- 9*x2

def f_(x2):
    return -f(x2)
```

```python
x0 = 0 # start from x = -5
xmin = fmin(f,x0)

x1 = -1 # start from x = -4
xmax = fmin(f_,x1)

x2 = np.linspace(-5.0, 5.0, 100)
# y = np.array([math.pow(x_i, 3) - (9*x_i) for x_i in x])
y2 = f(x2)
pylab.plot(x2,y2)
pylab.plot(x0, f(x0), 'bd', xmin, f(xmin), 'bo')
pylab.plot(x1, f(x1), 'rd', xmax, f(xmax), 'ro')
pylab.axis([-5,5,-30,30])
pylab.show()
```

```python
# c) f = -2 * sinx /x
x3 = np.linspace(-7.0, 7.0, 100)
y3 = -2 * np.sin(x3)/x3
y3min = -y3
x3max, _ = find_peaks(y3, height=0)
x3min, _ = find_peaks(y3min, height=0)
plt.plot(y3)
plt.plot(x3max, y3[x3max], "x")
plt.plot(x3min, y3[x3min], "p")
plt.show()
```

### Задание 4:

Градиентный спуск своими руками

```python
def f(x):
    return (x[0]**2) + (x[1]**2)
# derivative(f, x)
```

```python
x = np.array([100, 200])
lr = 0.1  
```

```python
gradient_dict = {}

for i in range(100):
    eps = np.sqrt(np.finfo(float).eps)
    gradient = approx_fprime(x, f, eps)
    gradient_dict[f(x)] = gradient
    print(f"Значении функции {f(x)} в точках  {gradient}")
    x = x - x * lr
```

```python
fmin = min(list(gradient_dict.keys()))
xmin = gradient_dict[min(list(gradient_dict.keys()))] 
```

```python
print("Минимум достигается в: ", xmin)
print("Значение функции в минимуме: ", fmin)
```

Проверка с помощью встроенной функции scipy: 

```python
minimize(f, x, method='nelder-mead', options={'xtol': 1e-8, 'disp': True})
```
