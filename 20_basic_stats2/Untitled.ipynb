```python
import numpy as np
import pandas as pd
import json
import matplotlib as plt
import re
```

```python
dict_string = {}
i = 0
with open('horse_data.names', "r", encoding='UTF-8') as f:
    for line in f:
        namespace = re.findall(r'([0-9]{1,2}):', line)
        if namespace:
            dict_string[namespace[0].strip()] = re.findall(r'[0-9]{1,2}:(.*)', line)[0].strip()
        i+=1
```

```python
df = pd.read_csv('horse_data.csv', names=range(0, 28))
data = df.drop(df.columns[[-2, -3]], axis='columns')
data.columns = dict_string.values()
data
```

```python
data.loc[data['surgery?'] == '1', 'surgery?'] = 'Yes'
data.loc[data['surgery?'] == '2', 'surgery?'] = 'No'

data.loc[data['Age'] == 1, 'Age'] = 'Adult'
data.loc[data['Age'] == 9, 'Age'] = 'Young'

data.loc[data['abdomen'] == '1', 'abdomen'] = 'normal'
data.loc[data['abdomen'] == '2', 'abdomen'] = 'other'
data.loc[data['abdomen'] == '3', 'abdomen'] = 'firm feces in the large intestine'
data.loc[data['abdomen'] == '4', 'abdomen'] = 'distended small intestine'
data.loc[data['abdomen'] == '5', 'abdomen'] = 'distended large intestine'

data.loc[data['mucous membranes'] == '1', 'mucous membranes'] = 'normal pink'
data.loc[data['mucous membranes'] == '2', 'mucous membranes'] = 'bright pink'
data.loc[data['mucous membranes'] == '3', 'mucous membranes'] = 'pale pink'
data.loc[data['mucous membranes'] == '4', 'mucous membranes'] = 'pale cyanotic'
data.loc[data['mucous membranes'] == '5', 'mucous membranes'] = 'bright red / injected'
data.loc[data['mucous membranes'] == '6', 'mucous membranes'] = 'dark cyanotic'

data.loc[data['outcome'] == '1', 'outcome'] = 'lived'
data.loc[data['outcome'] == '2', 'outcome'] = 'died'
data.loc[data['outcome'] == '3', 'outcome'] = 'was euthanized'
```

```python
data_filtered = data[['surgery?', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'abdomen', 'mucous membranes', 'outcome' ]]
```

```python
data_final = data_filtered.replace('?', np.NaN)
```

```python
data_final['rectal temperature'] = data_final['rectal temperature'].astype(object).astype(float)
data_final['pulse'] = data_final['pulse'].astype(object).astype(float)
data_final['respiratory rate'] = data_final['respiratory rate'].astype(object).astype(float)
```

Данные показывают, что температура тела особо не зависит от возраста и других категорий, разброс не большой, поэтому корректировка выбросов не требуется.
С пульсом и частотой дыхания сложнее, тут должна быть зависимость между собой. Проверим с помощью корреляции Пирсона:

```python
data_final.corr()
```

Как видим, есть связь между пульсом и частотой дыхания. Это надо будет учитывать в дальнейшем.

Работаем с выбросами. Только по значениям пульс и частота дыхания. Так как есть связь добавим дополнительный столбец pulse/respiratory и будем искать выбросы по этому столбцу. 

```python
data_final['pulse/respiratory'] = 1+data_final['pulse']/data_final['respiratory rate']
```

```python
q1 = data_final['pulse/respiratory'].quantile(0.25)
q3 = data_final['pulse/respiratory'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
data_normalized = data_final[data_final['pulse/respiratory'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse/respiratory')
print(data_final.describe())
print(data_normalized.describe())
print(lower_bound)
print(upper_bound)
data_normalized
```

```python
data_final.loc[(data_final['pulse/respiratory']>upper_bound)|(data_final['pulse/respiratory']<lower_bound)]
```

Видим, что выбросы по столбцу pulse/respiratory попали животные, которые умерли или были усыплены, что соответственно могло привести к нарушению частоты сердца и затрудненному дыханию.
Пропускать эти данные будет некорректно, поэтому будем использовать все данные.

Работа с пропусками

```python
data_final[data_final['outcome'].isnull()]
```

```python
data_final[data_final['surgery?'].isnull()]
```

Данные строки заменить на моду, т.к. нет информации об операции и исходе, все другие показатели в норме,то скорее всего просто растройство желудка, думаю, выжила и операция не потребовалась.

```python
data_final['outcome'].fillna('lived', inplace=True)
data_final['surgery?'].fillna('No', inplace=True)
```

Много пропусков столбца abdomen связано скорее всего с тем, что необходимости осмотра брюшной полости и жалоб не было.
Соотвественно нужно заменить на "normal"

```python
data_final[data_final['abdomen'].isnull()]
```

```python
data_final['abdomen'].fillna('normal', inplace=True)
```

```python
data_final[data_final['mucous membranes'].isnull()].tail()
```

mucous membranes заменим модой, т.к. взаимосвязь с другими данными не обнаружена.

```python
data_final['mucous membranes'].fillna(str(data_final['mucous membranes'].mode()), inplace=True)
```

rectal temperature заменим средней, т.к. разброс не большой

```python
data_final['rectal temperature'].fillna(data_final['rectal temperature'].mean(), inplace=True)
```

А pulse и respiratory rate заменим на медиану, т.к. разброс большой, в зависимости от возраста, от которого в основном зависят эти два значения.

```python
data_final['pulse'].fillna(data_final.groupby('Age').pulse.transform('median'), inplace=True)
```

```python
data_final['respiratory rate'].fillna(data_final.groupby('Age')['respiratory rate'].transform('median'), inplace=True)
```
