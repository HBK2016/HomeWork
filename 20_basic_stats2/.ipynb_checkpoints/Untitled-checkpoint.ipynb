```python
import numpy as np
import pandas as pd
import json
import matplotlib as plt
import re
```

```python
dict_string = {}
i = 0
with open('horse_data.names', "r", encoding='UTF-8') as f:
    for line in f:
        namespace = re.findall(r'([0-9]{1,2}):', line)
        if namespace:
            dict_string[namespace[0].strip()] = re.findall(r'[0-9]{1,2}:(.*)', line)[0].strip()
        i+=1
```

```python
df = pd.read_csv('horse_data.csv', names=range(0, 28))
data = df.drop(df.columns[[-2, -3]], axis='columns')
data.columns = dict_string.values()
data
```

```python
def fill_surgery(row):
    data_filtered['surgery?'] = data_filtered['surgery?'].str.replace('1', 'Yes')
    data_filtered['surgery?'] = data_filtered['surgery?'].str.replace('2', 'No')
    
def fill_Age(row):
    data_filtered['Age'] = data_filtered['Age'].replace(1, 'Adult')
    data_filtered['Age'] = data_filtered['Age'].replace(9, 'Young')
        
def fill_abdomen(row):
    data_filtered['abdomen'] = data_filtered['abdomen'].str.replace('1', 'normal')
    data_filtered['abdomen'] = data_filtered['abdomen'].str.replace('2', 'other')
    data_filtered['abdomen'] = data_filtered['abdomen'].str.replace('3', 'firm feces in the large intestine')
    data_filtered['abdomen'] = data_filtered['abdomen'].str.replace('4', 'distended small intestine')
    data_filtered['abdomen'] = data_filtered['abdomen'].str.replace('5', 'distended large intestine')
    
def fill_membranes(row):
    data_filtered['mucous membranes'] = data_filtered['mucous membranes'].str.replace('1', 'normal pink')
    data_filtered['mucous membranes'] = data_filtered['mucous membranes'].str.replace('2', 'bright pink')
    data_filtered['mucous membranes'] = data_filtered['mucous membranes'].str.replace('3', 'pale pink')
    data_filtered['mucous membranes'] = data_filtered['mucous membranes'].str.replace('4', 'pale cyanotic')
    data_filtered['mucous membranes'] = data_filtered['mucous membranes'].str.replace('5', 'bright red / injected')
    data_filtered['mucous membranes'] = data_filtered['mucous membranes'].str.replace('6', 'dark cyanotic')  
     
def fill_outcome(row): 
    data_filtered['outcome'] = data_filtered['outcome'].str.replace('1', 'lived')
    data_filtered['outcome'] = data_filtered['outcome'].str.replace('2', 'died')
    data_filtered['outcome'] = data_filtered['outcome'].str.replace('3', 'was euthanized')
   
```

```python
data_filtered = data[['surgery?', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'abdomen', 'mucous membranes', 'outcome' ]]
```

```python
data_filtered.is_copy = False
```

```python
# data_filtered.loc[data_filtered["surgery?"]=="2", "surgery?"] = "No"

data_filtered.apply(fill_surgery)
data_filtered.apply(fill_Age)
data_filtered.apply(fill_abdomen)
data_filtered.apply(fill_membranes)
data_filtered.apply(fill_outcome)
```

```python
data_final = data_filtered.replace('?', '0')

# data_filtered[].replace('?', np.nan)
# data_filtered['surgery?'].replace('?', data_filtered['surgery?'].mode())
# data_filtered['Age'].replace('?', data_filtered['Age'].mode())
# data_filtered['rectal temperature'].replace('?', data_filtered['rectal temperature'].median())
# data_final['pulse'].fillna(data_final['pulse'].median())
# data_final['respiratory rate'].fillna(data_final['respiratory rate'].median())
# data_final['abdomen'].fillna(data_final['abdomen'].mode())
# data_final['mucous membranes'].fillna(data_final['mucous membranes'].mode())
# data_final['outcome'].fillna(data_final['outcome'].mode())
```

```python
data_final['rectal temperature'] = data_final['rectal temperature'].astype(str).astype(float)
data_final['rectal temperature'] = data_final['pulse'].astype(str).astype(float)
data_final['rectal temperature'] = data_final['respiratory rate'].astype(str).astype(float)
```

```python
data_final.info()
```

```python
data_final.loc[data_final['rectal temperature'] == '0'] = data_final['rectal temperature'].median()
data_final.loc[data_final['surgery?'] == '0'] = data_final['surgery?'].mode()
data_final.loc[data_final['Age'] == '0'] = data_final['Age'].mode()
data_final.loc[data_final['pulse'] == '0'] = data_final['pulse'].median()
data_final.loc[data_final['respiratory rate'] == '0'] = data_final['respiratory rate'].median()
data_final.loc[data_final['abdomen'] == '0'] = data_final['abdomen'].mode()
data_final.loc[data_final['mucous membranes'] == '0'] = data_final['mucous membranes'].mode()
data_final.loc[data_final['outcome'] == '0'] = data_final['outcome'].mode()
```

```python
#Max
print(f"Максимальная температура тела {(data_final['rectal temperature'].max())}")
print(f"Максимальный пульс {data_final.pulse.max()}")
print(f"Максимальная частота дыхания {data_final['respiratory rate'].max()}")
```

```python
#Min
print(f"Минимальная температура тела {data_final['rectal temperature'].min()}")
print(f"Минимальная пульс {data_final.pulse.min()}")
print(f"Минимальная частота дыхания {data_final['respiratory rate'].min()}")
```

```python
#Размах
temper = float(data_final['rectal temperature'].max()) - float(data_final['rectal temperature'].min())
pulse = float(data_final.pulse.max()) - float(data_final.pulse.min())
respirat = float(data_final['respiratory rate'].max()) - float(data_final['respiratory rate'].min())


print(f"Размах температуры тела {temper}")
print(f"Размах пульса {pulse}")
print(f"Размах частоты дыхания {respirat}")
```

```python
#Среднеарифметическое

print(f"температура тела {data_final['rectal temperature'].mean()}")
print(f"пульс {data_final.pulse.mean()}")
print(f"частота дыхания {data_final['respiratory rate'].mean()}")
```

```python
#Мода
print(f"температура тела {data_final['rectal temperature'].mode()}")
print(f"пульс {data_final.pulse.mode()}")
print(f"частота дыхания {data_final['respiratory rate'].mode()}")
```

```python
#Медиана
print(f"температура тела {data_final['rectal temperature'].median()}")
print(f"пульс {data_final.pulse.median()}")
print(f"частота дыхания {data_final['respiratory rate'].median()}")
```
