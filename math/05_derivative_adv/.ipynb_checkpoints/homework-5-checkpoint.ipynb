### –ó–∞–¥–∞–Ω–∏–µ 2:

–ì—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫ —Å–≤–æ–∏–º–∏ —Ä—É–∫–∞–º–∏

```python
def f(x):
    return (x[0]**2) + (x[1]**2) 
```

```python
x = np.array([100, 200])
lr = 0.1
```

```python
gradient_dict = {}

for i in range(100):
    eps = np.sqrt(np.finfo(float).eps)
    gradient = approx_fprime(x, f, eps)
    gradient_dict[f(x)] = gradient
    print(f"–ó–Ω–∞—á–µ–Ω–∏–∏ —Ñ—É–Ω–∫—Ü–∏–∏ {f(x)} –≤ —Ç–æ—á–∫–∞—Ö  {gradient}")
    x = x - x * lr
```

```python
fmin = min(list(gradient_dict.keys()))
xmin = gradient_dict[min(list(gradient_dict.keys()))] 
```

```python
print("–ú–∏–Ω–∏–º—É–º –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è –≤: ", xmin)
print("–ó–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –≤ –º–∏–Ω–∏–º—É–º–µ: ", fmin)
```

–ü—Ä–æ–≤–µ—Ä–∫–∞ —Å –ø–æ–º–æ—â—å—é –≤—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ numpy: 

```python
minimize(f, x, method='nelder-mead', options={'xtol': 1e-8, 'disp': True})
```
