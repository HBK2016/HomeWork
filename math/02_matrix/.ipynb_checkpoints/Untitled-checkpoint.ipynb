При помощи матричных операций выполните:
Уменьшить объект в два раза.
Повернуть на 130 градусов.
Отразить объект относительно прямой y=x
Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1,2)

```python
A = [ [-100, -100, 1], [-100, 100, 1], [100, 100, 1], [0, 0, 1], [100, -100, 1], [-100, -100, 1] ]
A = np.array(A) 
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.show
```

```python
B = A / 2
B = np.array(B)
x1 = B[:,0]
y1 = B[:,1]
plt.plot(x1, y1)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.show
```

```python
theta = -130 * pi/180
TransformArray = np.array([
    [cos(theta), sin(theta), 0],
    [-sin(theta), cos(theta), 0],
    [0, 0, 1]
])

C = []
for point in A:
    new_point=np.dot(TransformArray,np.transpose(point))
    C.append(new_point)
C = np.array(C).astype(int)

x2 = C[:,0]
y2 = C[:,1]
plt.plot(x2, y2)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.show
```

```python
x0 = range(200, -201, -1)
y0 = x0
plt.plot(x0, y0)
trC = C + np.array([0, 200, 1])
xtr = trC[:,0]
ytr = trC[:,1]
plt.plot(xtr, ytr)
plt.ylim([-200, 500])
plt.xlim([-200, 200])
plt.show
# list(range(200, -201, -1))
```

```python
vector = np.array([200, 300, 1])

D = A + vector
D = np.array(D)
x3 = D[:,0]
y3 = D[:,1]
plt.plot(x3, y3)
plt.ylim([-100, 500])
plt.xlim([-100, 500])
plt.show
```

```python
vector2 = np.array([
    [0.5, 0, 0],
    [0, 1.2, 0],
    [0, 0, 1]
])

F = []
for point in A:
    new_point=np.dot(vector2,np.transpose(point))
    F.append(new_point)
F = np.array(F).astype(int)

x4 = F[:,0]
y4 = F[:,1]
plt.plot(x4, y4)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.show
```
